## 문제 8-2 module_usage.py

```python
import datetime
import random

# 현재 날짜 시간
now = datetime.datetime.now()
print("현재 날짜와 시간: %s" % (now.strftime("%Y-%m-%d %H:%M:%S")))
print("포맷된 날짜: %s" % (now.strftime("%Y년 %m월 %d일 %A")))

# 랜덤 값 생성
print("임의의 숫자: %d" % (random.randint(1, 10)))
print("임의의 실수: %.2f" % (random.uniform(1, 5)))

fruits = ["사과", "바나나", "오렌지", "포도", "딸기"]

print("임의의 리스트 요소: %s" % (random.choice(fruits)))

random.shuffle(fruits)
print("섞인 리스트: %s" % (fruits,))
```

## 문제 8-3 system_info.py

```python
import os
import sys

# 현재 작업 디렉토리
print("현재 작업 디렉토리: %s" % (os.getcwd()))

# Python 버전
print("Python 버전: %s" % (sys.version.split("\n")[0]))

# 운영체제
print("운영체제: %s" % (os.name))

# PATH 일부 출력
path_env = os.environ.get("PATH", "")
print("환경 변수 PATH 일부: %s" % (path_env[:30]))

# 파일 경로 처리
file_path = "/Users/username/documents/report.txt"

directory = os.path.dirname(file_path)
filename = os.path.basename(file_path)
extension = os.path.splitext(file_path)[1]

print("파일 경로 정보:")
print("- 디렉토리: %s" % (directory))
print("- 파일명: %s" % (filename))
print("- 확장자: %s" % (extension))

print("파일 존재 여부: %s" % (os.path.exists(file_path)))
```
