## 문제 6-1 calculator_functions.py

```python
x = int(input("첫 번째 숫자를 입력하세요: "))
y = int(input("두 번째 숫자를 입력하세요: "))

print("%d + %d = %d" % (x, y, x + y))
print("%d - %d = %d" % (x, y, x - y))
print("%d * %d = %d" % (x, y, x * y))
print("%d / %d = %.2f" % (x, y, x / y))
```

## 문제 6-2 factorial_functions.py

```python
def factorial_recursive(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial_recursive(n - 1)

def factorial_iterative(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print("5! (재귀) = %d" % (factorial_recursive(5)))
print("5! (반복) = %d" % (factorial_iterative(5)))
print("7! (재귀) = %d" % (factorial_recursive(7)))
print("7! (반복) = %d" % (factorial_iterative(7)))
```

## 문제 6-3 greeting_function.py

```python
def greet(name, language="ko", message=""):
    if language == "ko":
        greeting = "안녕하세요"
        if message != "":
            return "%s, %s님! %s" % (greeting, name, message)
        return "%s, %s님!" % (greeting, name)
    else:
        return "Hello, %s!" % (name)

print("%s" % (greet("김철수")))
print("%s" % (greet("John", language="en")))
print("%s" % (greet("이영희", message="좋은 하루 되세요!")))
```

## 문제 6-4 statistics_function.py

```python
import math

def calculate_statistics(numbers):
    mean = sum(numbers) / len(numbers)
    max_value = max(numbers)
    min_value = min(numbers)
    
    variance = sum((x - mean) ** 2 for x in numbers) / len(numbers)
    std_dev = math.sqrt(variance)
    
    return mean, max_value, min_value, std_dev

numbers = [10, 20, 30, 40, 50]
mean, max_value, min_value, std_dev = calculate_statistics(numbers)

print("숫자들: %s" % (numbers,))
print("평균: %.1f" % (mean))
print("최댓값: %d" % (max_value))
print("최솟값: %d" % (min_value))
print("표준편차: %.2f" % (std_dev))
```

## 문제 6-5 lambda_sorting.py

```python
students = [
    ("김철수", 85),
    ("이영희", 92),
    ("박민수", 78),
    ("최수진", 95)
]

name_sorted = sorted(students, key=lambda x: x[0])
score_sorted = sorted(students, key=lambda x: x[1])
score_desc = sorted(students, key=lambda x: x[1], reverse=True)

print("학생 정보: %s" % (students,))
print("이름순 정렬: %s" % (name_sorted,))
print("점수순 정렬: %s" % (score_sorted,))
print("점수 내림차순: %s" % (score_desc,))
```
