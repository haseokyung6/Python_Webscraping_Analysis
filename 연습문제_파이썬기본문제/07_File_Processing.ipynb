## 문제 7-1 file_write_read.py

```python
content = """안녕하세요
파이썬 파일 처리를 연습하고 있습니다
오늘은 좋은 날씨입니다
"""

# 파일에 쓰기
with open("7-1.txt", "w", encoding="utf-8") as file:
    file.write(content)

print("파일에 저장할 내용:")
print(content)

# 파일에서 읽기
print("파일에서 읽어온 내용:")
with open("7-1.txt", "r", encoding="utf-8") as file:
    read_content = file.read()
    print(read_content)
```

## 문제 7-2 csv_grade_processor.py

```python
students = [
    ("김철수", 85),
    ("이영희", 92),
    ("박민수", 78),
    ("최수진", 95)
]

# CSV 파일 저장
with open("grades.csv", "w", encoding="utf-8") as file:
    for name, score in students:
        file.write("%s,%d\n" % (name, score))

print("학생 성적이 grades.csv에 저장되었습니다.\n")

# 파일 읽기 및 평균 계산
total = 0
count = 0

print("성적 분석 결과:")
with open("grades.csv", "r", encoding="utf-8") as file:
    for line in file:
        name, score = line.strip().split(",")
        score = int(score)
        print("%s: %d점" % (name, score))
        total += score
        count += 1

average = total / count
print("전체 평균: %.1f점" % (average))
```

## 문제 7-3 word_frequency.py

```python
text = """파이썬 프로그래밍 언어
파이썬 언어 배우기
쉬운 프로그래밍
강력한 파이썬"""

# 파일 저장
with open("7-3.txt", "w", encoding="utf-8") as file:
    file.write(text)

# 단어 빈도 계산
word_count = {}

with open("7-3.txt", "r", encoding="utf-8") as file:
    for line in file:
        words = line.strip().split()
        for word in words:
            word_count[word] = word_count.get(word, 0) + 1

print("단어 빈도 분석 결과:")
for word, count in word_count.items():
    print("%s: %d번" % (word, count))
```

## 문제 7-4 json_handler.py

```python
import json

data = {
    "이름": "김철수",
    "나이": 25,
    "직업": "개발자",
    "취미": ["독서", "영화감상", "코딩"],
    "주소": "서울시 강남구"
}

# JSON 저장
with open("data.json", "w", encoding="utf-8") as file:
    json.dump(data, file, ensure_ascii=False, indent=4)

print("데이터가 data.json에 저장되었습니다.\n")

# JSON 읽기
with open("data.json", "r", encoding="utf-8") as file:
    loaded_data = json.load(file)

print("JSON에서 읽어온 데이터:")
for key, value in loaded_data.items():
    print("%s: %s" % (key, value))
```

## 문제 7-5 log_processor.py

```python
logs = [
    "2025-07-20 09:15:00 - WARNING - 메모리 사용량이 높습니다",
    "2025-07-20 10:30:00 - ERROR - 데이터베이스 연결 실패",
    "2025-07-20 11:45:00 - ERROR - 파일을 찾을 수 없음",
    "2025-07-20 12:00:00 - WARNING - 디스크 공간 부족"
]

# 로그 파일 생성
with open("app.log", "w", encoding="utf-8") as file:
    for log in logs:
        file.write(log + "\n")

print("로그 파일이 생성되었습니다.\n")

# 특정 레벨 필터링
error_logs = []
warning_logs = []

with open("app.log", "r", encoding="utf-8") as file:
    for line in file:
        if "ERROR" in line:
            error_logs.append(line.strip())
        elif "WARNING" in line:
            warning_logs.append(line.strip())

print("ERROR 레벨 로그들:")
for log in error_logs:
    print(log)

print("\nWARNING 레벨 로그들:")
for log in warning_logs:
    print(log)
```
