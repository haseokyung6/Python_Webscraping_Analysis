## 문제 3-1 list_statistics.py

```python
numbers= [10, 20, 30, 40, 50]

total = sum(numbers)
average = total / len(numbers)

print("숫자들: %s" % (numbers))
print("합계: %d" % total)
print("평균: %.1f" % average)
```

## 문제 3-3 student_grades.py

```python
grades = {
    "김철수": 85,
    "이영희": 92,
    "박민수": 78,
    "최수진": 95
}
print("학생 성적:")
total=0
for name in grades:
    print("%s: %d점" % (name, grades[name]))
    total += grades[name]

average = total / len(grades)
print("평균 점수: %.1f점" % average)
```

## 문제 3-4 distance_calculator.py

```python
import math

point1 = (0,0)
point2 = (3,4)

distance = math.sqrt((point2[0] - point1[0])**2 + (point2[1] - point1[1])**2)

print("점1: %s" % (point1, ))
print("점2: %s" % (point2, ))
print("두 점 사이의 거리: %.1f" % (distance))
```

## 문제 3-5 shopping_cart.py

```python
cart = {
    "사과": (2, 1000),
    "바나나": (3, 800),
    "오렌지": (1, 1500)
}

print("쇼핑 카트:")

total_price = 0

for item in cart:
    quantity = cart[item][0]
    price = cart[item][1]
    item_total = quantity * price
    total_price += item_total
    
    print("%s: %d개 (개당 %d원) = %d원" % (item, quantity, price, item_total))

print("총 가격: %d원" % (total_price))
```

## 문제 3-6 list_analysis.py

```python
numbers = [15, 3, 27, 8, 19, 12, 31]
max_num = max(numbers)
min_num = min(numbers)

sorted_numbers = sorted(numbers, reverse=True)
second_max = sorted_numbers[1]

print("숫자 목록: %s" % (numbers))
print("최대값: %d" % max_num)
print("최소값: %d" % min_num)
print("두 번째로 큰 값: %d" % second_max)
```

## 문제 3-7 word_length.py

```python
words = ['cat', 'elephant', 'dog', 'butterfly', 'ant']

longest = max(words, key=len)
shortest = min(words, key=len)

print("단어 목록: %s" % (words))
print("가장 긴 단어: %s (%d글자)" % (longest, len(longest)))
print("가장 짧은 단어: %s (%d글자)" % (shortest, len(shortest)))
```

## 문제 3-8 remove_duplicates.py

```python
numbers = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3]

unique_numbers = list(set(numbers))
unique_numbers.sort()

print("원본 리스트: %s" % (numbers, ))
print("중복 제거된 리스트: %s" % (unique_numbers, ))
```

## 문제 3-9 list_operations.py

```python
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]

merged = sorted(list(set(list1 + list2)))
common = sorted(list(set(list1) & set(list2)))

print("리스트1: %s" % (list1,))
print("리스트2: %s" % (list2,))
print("병합된 리스트: %s" % (merged,))
print("공통 요소: %s" % (common,))
```

## 문제 3-10 student_sort.py

```python
students = [
    {"이름": "김철수", "나이": 20, "학과": "컴퓨터공학과"},
    {"이름": "이영희", "나이": 22, "학과": "영어영문학과"},
    {"이름": "박민수", "나이": 21, "학과": "경영학과"},
    {"이름": "최수진", "나이": 23, "학과": "수학과"}
]

sorted_students = sorted(students, key=lambda x: x["나이"])
print("나이 순으로 정렬된 학생 목록:")
for student in sorted_students:
    print("%s (%d세) - %s" % (student["이름"], student["나이"], student["학과"]))
```
