## 문제 1-1 basic_calculator.py

```python
x = int(input("첫 번째 숫자를 입력하세요: "))
y = int(input("두 번째 숫자를 입력하세요: "))

print("%d + %d = %d" % (x, y, x + y))
print("%d - %d = %d" % (x, y, x - y))
print("%d * %d = %d" % (x, y, x * y))
print("%d / %d = %.2f" % (x, y, x / y))
```

## 문제 1-2  circle_calculator.py

```python
pi = 3.14159

r = float(input("원의 반지름을 입력하세요: "))

area = pi * r * r
circumference = 2 * pi * r

print("반지름이 %d인 원의 넓이: %.2f" % (r, area))
print("반지름이 %d인 원의 둘레: %.2f" % (r, circumference))
```

## 문제 1-3 temperature_converter.py

```python
c = int(input("섭씨 온도를 입력하세요: "))

f = c * 9 / 5 + 32

print("섭씨 %d도는 화씨 %.1f도입니다." % (c, f))
```

## 문제 1-4 rectangle_calculator.py

```python
x = int(input("가로 길이를 입력하세요: "))
y = int(input("세로 길이를 입력하세요: "))

area = x * y
perimeter = 2 * (x + y)

print("직사각형의 넓이: %d" % area)
print("직사각형의 둘레: %d" % perimeter)
```

## 문제 1-5 discount_calculator.py

```python
price = int(input("상품의 가격을 입력하세요: "))
discount = int(input("할인율을 입력하세요(%): "))

discount_amount = price * discount / 100
final_price = price - discount_amount

print("원래 가격: %d원" % price)
print("할인율: %d%%" % discount)
print("할인 금액: %d원" % discount_amount)
print("최종 가격: %d원" % final_price)
```
