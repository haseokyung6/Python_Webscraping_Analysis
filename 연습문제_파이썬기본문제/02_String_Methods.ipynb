## 문제 2-1 self_introduction.py

```python
name = input("이름을 입력하세요: ")
age = int(input("나이를 입력하세요: "))

print("안녕하세요! 제 이름은 %s이고, 나이는 %d살입니다." % (name, age))
```

## 문제 2-2 string_operations.py

```python
s = input("문자열을 입력하세요: ")

print("대문자: %s" % s.upper())
print("소문자: %s" % s.lower())
print("문자열 길이: %d" % len(s))
```

## 문제 2-3 email_parser.py

```python
s = input("이메일 주소를 입력하세요: ")

user_name, domain = s.split("@")
print("사용자명: %s" % user_name)
print("도메인: %s" % domain)
```

## 문제 2-4 word_counter.py

```python
sentence = input("문장을 입력하세요: ")

clean_sentence = " ".join(sentence.split())
word_count = len(clean_sentence.split())

print("공백 제거: %s" % (clean_sentence))
print("단어 개수: %d개" % (word_count))
```

## 문제 2-5 character_counter.py

```python
s = input("문자열을 입력하세요: ")
ch = input("찾을 문자를 입력하세요: ")

count = s.count(ch)

print("문자 '%s'이 %d번 나타납니다." % (ch, count))
```
