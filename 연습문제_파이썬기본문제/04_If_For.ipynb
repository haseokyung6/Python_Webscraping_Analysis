## 문제 4-1 grade-calculator.py

```python
x = int(input("점수를 입력하세요: "))

if x >= 90:
    print("점수 %d점의 학점은 A입니다." % x)
elif x >= 80:
    print("점수 %d점의 학점은 B입니다." % x)
elif x >= 70:
    print("점수 %d점의 학점은 C입니다." % x)
elif x >= 60:
    print("점수 %d점의 학점은 D입니다." % x)
else:
    print("점수 %d점의 학점은 F입니다." % x)
```

## 문제 4-2 multiple_sum.py

```python
multiples = []

for i in range(1, 101):
    if i % 3 == 0:
        multiples.append(i)

total = sum(multiples)
count = len(multiples)

print("1부터 100까지 3의 배수: %s" % (multiples,))
print("3의 배수의 합: %d" % (total))
print("3의 배수의 개수: %d개" % (count))
```

## 문제 4-3 multiplication_table.py

```python
dan = int(input("몇 단을 출력할까요? "))

print("%d단 구구단:" % (dan))

for i in range(1, 10):
    print("%d x %d = %d" % (dan, i, dan * i))
```

## 문제 4-4 number_sum.py

```python
total = 0

while True:
    num = int(input("숫자를 입력하세요 (0을 입력하면 종료): "))
    
    if num == 0:
        break
    
    total += num

print("입력된 숫자들의 합: %d" % (total))
```

## 문제 4-5 prime_checker.py

```python
num = int(input("숫자를 입력하세요: "))

is_prime = True

if num < 2:
    is_prime = False
else:
    for i in range(2,num):
        if num % i == 0:
            is_prime = False
            break

if is_prime:
    print("%d는 소수입니다." % (num))
else:    
    print("%d는 소수가 아닙니다." % (num))
```
