```python
from urllib.parse import urljoin
from IPython.display import Image, display
import requests
from bs4 import BeautifulSoup

# 요청 헤더 설정 : 브라우저 정보
req_header = {
    'user-agent': (
        'Mozilla/5.0 (Windows NT 10.0; Win64; x64) '
        'AppleWebKit/537.36 (KHTML, like Gecko) '
        'Chrome/104.0.0.0 Safari/537.36'
    )
}

def scrape_nate_sections(max_per_section=5):
    base = 'https://news.nate.com/recent?mid='
    sections = {
        '최신뉴스': 'n0100',
        '정치': 'n0101',
        '경제': 'n0102',
        '사회': 'n0103',
        '세계': 'n0104',
        'IT/과학': 'n0105',
    }

    for name, mid in sections.items():
        url = base + mid
        print(f'--- {name} ({url}) ---')
        try:
            r = requests.get(url, headers=req_header, timeout=10)
            r.raise_for_status()
            # Ensure proper encoding for euc-kr pages
            if not r.encoding or r.encoding.lower() in ('iso-8859-1', 'windows-1252'):
                r.encoding = 'euc-kr'
        except Exception as e:
            print('Request failed:', e)
            continue

        soup = BeautifulSoup(r.text, 'html.parser')

        items = soup.select('div.mduSubjectList div.mlt01')
        if not items:
            print('No articles found for', name)
            continue

        for idx, item in enumerate(items, 1):
            if idx > max_per_section:
                break

            a = item.select_one('a.lt1')
            if not a:
                continue

            href = a.get('href')
            full_href = urljoin(url, href) if href else None

            # title
            h2 = a.select_one('h2.tit')
            title = h2.get_text(strip=True) if h2 else a.get_text(strip=True)

            # image (check existence)
            img = a.select_one('em.mediatype img') or a.select_one('img')
            img_url = None
            if img:
                src = img.get('src') or img.get('data-src') or ''
                src = src.strip()
                if src:
                    img_url = urljoin(url, src)

            # display results
            if img_url:
                try:
                    display(Image(url=img_url))
                except Exception:
                    pass

            print(title)
            print(full_href)            
            print()

    print('Done.')


scrape_nate_sections()
```

```python
import os
import requests
from urllib.parse import urljoin
from bs4 import BeautifulSoup

def download_one_episode(title, no, url):
    # 디렉토리 생성: img/<title>/<no>
    base_dir = os.path.join('img', str(title), str(no))
    os.makedirs(base_dir, exist_ok=True)

    # 요청 헤더: referer는 에피소드 페이지 URL로 설정
    req_header = {
        'referer': url,
        'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) '
                      'AppleWebKit/537.36 (KHTML, like Gecko) '
                      'Chrome/104.0.0.0 Safari/537.36'
    }

    try:
        res = requests.get(url, headers=req_header, timeout=10)
        res.raise_for_status()
    except Exception as e:
        print('Failed to fetch episode page:', e)
        return

    # 파싱 (네이버 웹툰은 일반적으로 img 태그에 직접 이미지가 있음)
    soup = BeautifulSoup(res.text, 'html.parser')

    # 우선적으로 웹툰 이미지가 들어있는 뷰어 영역을 선택
    selectors = [
        'div.wt_viewer img',
        'div.view_wrap img',
        'div#comic_viewer img',
        'div.comic_viewer img',
    ]
    img_urls = []
    for sel in selectors:
        imgs = soup.select(sel)
        if imgs:
            for img in imgs:
                src = img.get('src') or img.get('data-src') or ''
                if src:
                    img_urls.append(urljoin(url, src.strip()))
            if img_urls:
                break

    # fallback: 모든 img 태그 검사 (필요 시 필터링)
    if not img_urls:
        for img in soup.find_all('img'):
            src = img.get('src') or img.get('data-src') or ''
            if not src:
                continue
            full = urljoin(url, src.strip())
            # pstatic 등 웹툰 이미지 도메인만 포함시키는 필터 - 필요에 따라 조정
            if 'pstatic' in full or 'image' in full:
                img_urls.append(full)

    if not img_urls:
        print('No images found on the episode page.')
        return

    # 다운로드
    for idx, img_url in enumerate(img_urls, start=1):
        try:
            r = requests.get(img_url, headers=req_header, timeout=15)
            if not r.ok:
                print('Failed to download', img_url)
                continue
            img_data = r.content
            file_name = os.path.basename(img_url.split('?')[0])
            out_path = os.path.join(base_dir, f'{idx:03d}_' + file_name)
            with open(out_path, 'wb') as f:
                f.write(img_data)
            print(f'Writing to {out_path} ({len(img_data):,} bytes)')
        except Exception as e:
            print('Error saving', img_url, e)

    print('Download complete.')

download_one_episode('일렉시드',341,'https://comic.naver.com/webtoon/detail?titleId=717481&no=341&week=wed')
```
