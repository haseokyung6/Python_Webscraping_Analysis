```python
# reqeusts, bs4 import
import requests
import bs4
# BeautifulSoup 클래스 import
from bs4 import BeautifulSoup
```

```python
# requests, bs4 버전 확인하기
print(f'requests 버전 {requests.__version__}')
print(f'beautifulsoup4 버전 {bs4.__version__}')
```

```python
# 다음 경제 뉴스
url = 'https://news.daum.net/economy'

# 요청 헤더 설정 : 브라우저 정보
req_header = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36"
}

# requests의 get() 함수 호출하기
res = requests.get(url, headers=req_header)
print(res)
print(type(res))
print(res.status_code, res.ok)

# 응답(response)이 OK 이면
if res.ok:
    # 응답(response)을 utf-8로 encoding
    res.encoding = 'utf-8'
    html = res.text
    
    # BeautifulSoup 객체 생성
    soup = BeautifulSoup(html, "html.parser")
    
    # ul.list_newsheadline2 li 선택
    for li_tag in soup.select('ul.list_newsheadline2 li'):
        # a 태그에서 링크 추출
        a_tag = li_tag.find('a')
        if a_tag:
            link = a_tag['href']
            
            # div.cont_thumb strong.tit_txt에서 제목 추출
            strong_tag = li_tag.select_one('div.cont_thumb strong.tit_txt')
            if strong_tag:
                title = strong_tag.text.strip()
                print(f"{link}\n{title}\n")
else:
    # 응답(response)이 Error 이면 status code 출력
    print(f"에러 코드 = {res.status_code}")
```

```python
section_dict = {'기후/환경':'climate','사회':'society','경제':'economy','정치':'politics',\
             '국제':'world','문화':'culture','생활':'life','IT/과학':'tech','인물':'people'}

# 함수선언
def print_news(section_name):
    # section_name에서 영문 코드 추출
    section_code = section_dict[section_name]
    # URL 생성
    url = f'https://news.daum.net/{section_code}'
    
    # 요청 헤더 설정 : 브라우저 정보
    req_header = {
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36"
    }
    
    # requests의 get() 함수 호출하기
    res = requests.get(url, headers=req_header)
    print(f"=== {section_name} 뉴스 ===")
    print(res.status_code, res.ok)
    
    # 응답(response)이 OK 이면
    if res.ok:
        # 응답(response)을 utf-8로 encoding
        res.encoding = 'utf-8'
        html = res.text
        
        # BeautifulSoup 객체 생성
        soup = BeautifulSoup(html, "html.parser")
        
        # ul.list_newsheadline2 li 선택
        for li_tag in soup.select('ul.list_newsheadline2 li'):
            # a 태그에서 링크 추출
            a_tag = li_tag.find('a')
            if a_tag:
                link = a_tag['href']
                
                # div.cont_thumb strong.tit_txt에서 제목 추출
                strong_tag = li_tag.select_one('div.cont_thumb strong.tit_txt')
                if strong_tag:
                    title = strong_tag.text.strip()
                    print(f"{link}\n{title}\n")
    else:
        # 응답(response)이 Error 이면 status code 출력
        print(f"에러 코드 = {res.status_code}")

# 함수호출
print_news('경제')
print_news('사회')
```
